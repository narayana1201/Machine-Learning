```python
# Get all unique values in the 'education' column
unique_values = data['education'].unique()
print(unique_values)
```

```python
# Keep only rows with unique values in 'education'
unique_data = first_100.drop_duplicates(subset='education')
unique_data.to_csv('unique_id.csv', index=False)
```

```python
# Filter rows where 'columname' column equals value
result = data[data['age'] == 27]
print(result)
```

```python
# Count occurrences of 'value' in the 'columname' column
count_40 = (data['purchase_amount'] == 2395).sum()
print(count_40)
```

```python
# Count occurrences of each value in the 'columname' column
amount_counts = data['purchase_amount'].value_counts()

# Display the count of '5819' in the 'amount' column
count = amount_counts.get(5819, 0)  # Getting the count for value '5819', defaulting to 0 if it's not found
print(count)
```

```python
# Count occurrences of each value in the 'columname' column
id = data['loyalty_status'].value_counts()
print(id)
```

```python
# detect missing values 
result = id.isnull() 
  
# Print the result 
print(result) 
```

```python
# Filtering rows where values in column 'Age' are greater than 30
filtered_single_value = first_100[first_100['age'] > 30]
print(filtered_single_value)
```

```python
# Filtering rows where values in column 'promotion_usage' match specific values
filtered_multiple_values = first_100[first_100['promotion_usage'].isin(['1', '0'])]
print(filtered_multiple_values)
```

```python
# Sorting DataFrame by values in column 'age' in ascending order
sorted_values = first_100.sort_values(by='age')

# Sorting DataFrame by index in descending order
#sorted_index = data.sort_index(ascending=False)

print(sorted_values)
```

```python
# class v / s fare barplot 
sns.barplot(x = 'age', y = 'income', data = first_100)
 
# Show the plot
plt.show()
```

```python
# Creating the new DataFrame and adding column and values
df_test = pd.DataFrame({'Weight': [45, 88, 56, 15, 71],
                   'Name': ['Sam', 'Andrea', 'Alex', 'Robin', 'Kia'],
                   'Age': [14, 25, 55, 8, 21]})
# Create the index
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']
 
# Set the index
df_test.index = index_

result = df_test.loc['Row_2', 'Name']
print("\nSelected Value at Row_2, Column 'Name':")
print(result)
```

```python
# Select a single value at a specific row and column location
value_at_index = df_test.iloc[2, 1]  # Row index 2, Column index 1
print("Value at row index 2, column index 1:", value_at_index)
```

```python
# Select multiple values at specific row and column locations
multiple_values_indices = data.iloc[[0, 2], [0, 2]]  # Rows 0 and 2, Columns 0 and 2
print("Multiple values at row indices 0 and 2, column indices 0 and 2:\n", multiple_values_indices)
```
